```python
from imdb import IMDb

ia = IMDb()

class Movie:
    def __init__(self, name, director, genre, year, rating):
        self.name = name
        self.director = director
        self.genre = genre
        self.year = year
        self.rating = rating

movies = []

from imdb import IMDb

def add_movie():
    print("Enter movie details: ")
    title = input("Title: ")

    imdb = IMDb()
    movies = imdb.search_movie(title)
    
    if not movies:
        print(f"No results found for {title}.")
        return

    print("Select a movie:")
    for i, movie in enumerate(movies, start=1):
        print(f"{i}. {movie['title']} ({movie['year']})")
    
    choice = int(input("Enter your choice: "))
    selected_movie = imdb.get_movie(movies[choice - 1].getID())
    
    name = selected_movie['title']
    director = ', '.join([d['name'] for d in selected_movie['directors']])
    genre = ', '.join(selected_movie['genres'])
    year = selected_movie['year']
    rating = selected_movie['rating']

    movie = Movie(name, director, genre, year, rating)
    movies.append(movie)
    print(f"{name} has been added to the list.")

    
    search_results = ia.search_movie(name)
    if not search_results:
        print("Movie not found on IMDb.")
        return
    movie_id = search_results[0].movieID
    movie = ia.get_movie(movie_id)
    director = ", ".join([person['name'] for person in movie['director']])
    genre = ", ".join(movie['genres'])
    year = str(movie['year'])
    rating = str(movie['rating'])
    
    movie = Movie(name, director, genre, year, rating)
    movies.append(movie)

def remove_movie():
    name = input("Enter movie name to remove: ")
    for movie in movies:
        if movie.name == name:
            movies.remove(movie)
            print(f"{name} has been removed.")
            break
    else:
        print(f"{name} not found.")

def list_movies():
    for movie in movies:
        print(f"{movie.name} ({movie.year}) - {movie.genre} - Directed by {movie.director} - Rating: {movie.rating}")

while True:
    print("Select an operation: ")
    print("1. Add movie")
    print("2. Remove movie")
    print("3. List movies")
    print("0. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_movie()
    elif choice == "2":
        remove_movie()
    elif choice == "3":
        list_movies()
    elif choice == "0":
        break
    else:
        print("Invalid choice. Try again.")
        
```
