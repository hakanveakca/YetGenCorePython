```python
class Movie:
    def __init__(self, name, director, genre, year, rating, log):
        self.name = name
        self.director = director
        self.genre = genre
        self.year = year
        self.rating = rating
        self.log = log

class Series(Movie):
    def __init__(self, name, director, genre, year, rating, log, seasons):
        super().__init__(name, director, genre, year, rating, log)
        self.seasons = seasons

movies = []
tvSeries = []
favorites = []
data_file = "data.txt"  

def save_data():
    with open(data_file, "w") as file:  
        for movie in movies:
            file.write(f"Movie|{movie.name}|{movie.director}|{movie.genre}|{movie.year}|{movie.rating}|{movie.log}\n")
        for series in tvSeries:
            file.write(f"Series|{series.name}|{series.director}|{series.genre}|{series.year}|{series.rating}|{series.log}|{series.seasons}\n")
        for content in favorites:
            if isinstance(content, Movie):
                file.write(f"Favorite|Movie|{content.name}|{content.director}|{content.genre}|{content.year}|{content.rating}|{content.log}\n")
            elif isinstance(content, Series):
                file.write(f"Favorite|Series|{content.name}|{content.director}|{content.genre}|{content.year}|{content.rating}|{content.log}|{content.seasons}\n")


def load_data():
    try:
        with open(data_file, "r") as file:
            for line in file:
                data = line.strip().split("|")
                if data[0] == "Movie":
                    movie = Movie(data[1], data[2], data[3], int(data[4]), float(data[5]), data[6])
                    movies.append(movie)
                elif data[0] == "Series":
                    series = Series(data[1], data[2], data[3], int(data[4]), float(data[5]), data[6], data[7])
                    tvSeries.append(series)
                elif data[0] == "Favorite":
                    if data[1] == "Movie":
                        movie = Movie(data[2], data[3], data[4], int(data[5]), float(data[6]), data[7])
                        favorites.append(movie)
                    elif data[1] == "Series":
                        series = Series(data[2], data[3], data[4], int(data[5]), float(data[6]), data[7], data[8])
                        favorites.append(series)
    except FileNotFoundError:
        pass

def add_movie():
    print("Enter movie details:")
    name = input("Name: ").capitalize()
    director = input("Director: ").capitalize()
    genre = input("Genre (Action, Horror, Romance, Drama, Animation, Documentary, Comedy): ").capitalize()
    year = int(input("Year: "))
    rating = float(input("Rating (0-10): "))
    log = input("Log: ").capitalize()

    movie = Movie(name, director, genre, year, rating, log)
    movies.append(movie)

    isFavorite = input("Do you want to add this movie to favorites (y/n): ")
    if isFavorite == "y":
        favorites.append(movie)

    save_data()
    print(f"{name} is added!")

def remove_movie():
    name = input("Enter the movie name you want to remove: ").capitalize()
    for movie in movies:
        if movie.name == name:
            movies.remove(movie)
            save_data()
            print(f"{name} has been removed.")
            break
    else:
        print(f"{name} not found.")

def list_movies():
    load_data()

    if len(movies) == 0:
        print("Movies are not found!")

    for movie in movies:
        print(f"{movie.name} ({movie.year}) - {movie.genre} - Directed by {movie.director} - Rating: {movie.rating} - Log: {movie.log}")

def add_series():
    print("Enter series details:")
    name = input("Name: ").capitalize()
    director = input("Director: ").capitalize()
    genre = input("Genre (Action, Horror, Romance, Drama, Animation, Documentary, Comedy): ").capitalize()
    year = int(input("Year: "))
    rating = float(input("Rating (0-10): "))
    log = input("Log: ").capitalize()
    seasons = input
    year = int(input("Year: "))
    rating = float(input("Rating (0-10): "))
    log = input("Log: ").capitalize()
    seasons = input("Seasons: ").capitalize()

    series = Series(name, director, genre, year, rating, log, seasons)
    tvSeries.append(series)

    isFavorite = input("Do you want to add this series to favorites (y/n): ")
    if isFavorite == "y":
        favorites.append(series)

    save_data()
    print(f"{name} is added!")
def remove_series():
    name = input("Enter the series name you want to remove: ").capitalize()
    for series in tvSeries:
        if series.name == name:
            tvSeries.remove(series)
            save_data()
            print(f"{name} has been removed.")
            break
    else:
        print(f"{name} not found.")

def list_series():
    load_data()
    if len(tvSeries) == 0:
        print("Series are not found!")

    for series in tvSeries:
        print(f"{series.name} ({series.year}) - {series.genre} - Directed by {series.director} - Rating: {series.rating} - Seasons: {series.seasons} - Log: {series.log}")

def search_movie():
    load_data()
    text = input("Enter the name you want to search for: ").capitalize()
    isFound = False
    count = 1
    for movie in movies:
        if text in movie.name:
            print(f"{count}. {movie.name}")
            count += 1
            isFound = True
    if not isFound:
        print(f"No movie containing '{text}' found!")

def search_series():
    load_data()
    text = input("Enter the name you want to search for: ").capitalize()
    isFound = False
    count = 1
    for series in tvSeries:
        if text in series.name:
            print(f"{count}. {series.name}")
            count += 1
            isFound = True
    if not isFound:
        print(f"No series containing '{text}' found!")

def list_favorites():
    load_data()
    if len(favorites) == 0:
        print("You have not added any content to favorites!")

    for content in favorites:
        if isinstance(content, Movie):
            print(f"{content.name} ({content.year}) - {content.genre} - Directed by {content.director} - Rating: {content.rating} - Log: {content.log}")
        elif isinstance(content, Series):
            print(f"{content.name} ({content.year}) - {content.genre} - Directed by {content.director} - Rating: {content.rating} - Seasons: {content.seasons} - Log: {content.log}")

def remove_from_favorites():
    load_data()
    name = input("Enter the name of the content you want to remove from favorites: ").capitalize()
    removed = False
    for content in favorites:
        if content.name == name:
            favorites.remove(content)
            save_data()
            print(f"{name} is removed from favorites!")
            removed = True
            break
    if not removed:
        print(f"{name} is not found in favorites!")

def search_for_genre(genre):
    load_data()
    found = False
    for movie in movies:
        if movie.genre == genre:
            print(movie.name)
            found = True

    for series in tvSeries:
        if series.genre == genre:
            print(series.name)
            found = True
    
    if not found:
        print(f"No content found in the genre '{genre}'!")

def sort_by_rating():
    load_data()
    sorted_content = sorted(movies + tvSeries, key=lambda x: x.rating, reverse=True)
    for content in sorted_content:
        if isinstance(content, Movie):
            print(f"{content.name} by {content.director} - Rating: {content.rating}")
        elif isinstance(content, Series):
            print(f"{content.name} by {content.director} - Rating: {content.rating} - Seasons: {content.seasons}")

print("""
Select an operation:
0. Exit 
1. Add movie
2. Remove movie
3. List movies
4. Add Series 
5. Remove Series
6. List Series
7. Search for movie
8. Search for series
9. List favorites
10. Remove from favorites
11. Search for genre
12. Sort by rating
""")


while True:
    choice = input("Enter your choice: ")

    if choice == "0":
        break
    elif choice == "1":
        add_movie()
    elif choice == "2":
        remove_movie()
    elif choice == "3":
        list_movies()
    elif choice == "4":
        add_series()
    elif choice == "5":
        remove_series()
    elif choice == "6":
        list_series()
    elif choice == "7":
        search_movie()
    elif choice == "8":
        search_series()
    elif choice == "9":
        list_favorites()
    elif choice == "10":
        remove_from_favorites()
    elif choice == "11":
        search_for_genre()
    elif choice == "12":
        sort_by_rating()
    else:
        print("Invalid choice. Try again.")
#fix
```
